### It is in the object type, we need to change it into the  int type 

```python
df['Installs'] = df['Installs'].astype(str)  # Ensure it's a string
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)  # '+' hatao
df['Installs'] = df['Installs'].str.replace(',', '', regex=True)   # ',' hatao
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')  # Convert safely to number
df['Installs'] = df['Installs'].astype('Int64')  # Use Pandas' Int64
```

### The size of the data can be i  GB, MB & KB let's convert the data in to size in MB

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else str(x))

```

```python
#df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('K', '')) / 1000 if 'K' in str(x) else x)

```

### Here we get missmatched value with the data we got the value 1,018 we can dropit or we can assume as it may be '.'(dot) they would incorrectly added to the database. So let assume it as dot for now and replace the ',' with the dot'.' 

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace(',', '.') if ',' in str(x) else x)
```

## Conversion of kbs to mb

```python
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
```

```python
#df['Size'] = df['Size'].apply(lambda x:float(x))
```

### we got another  missmatched value with the dataset which is 'Varies with device ' it means the size of App may varies with the device ----->but here assuming it as 0

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', '0') if 'Varies with device' in str(x) else (x))
```

```python
#df['Size'] = df['Size'].apply(lambda x:float(x))
```

## Conversion of data Gbs to Mb

```python
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('G', '')) / 1000 if 'G' in str(x) else x)
```

```python
df['Size'] = df['Size'].apply(lambda x: float(x))
```
